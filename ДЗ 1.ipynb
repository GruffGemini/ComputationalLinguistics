```python
import re

def remove_tags(txt:str) -> str:
    clean_text = re.sub(r'<.*?>', '', txt) # убираем теги вместе с содержимым
    clean_text = re.sub(r'([.!?])(\w)', r'\1 \2', clean_text)
    # вставляем пробел между знаком препинания и последующим знаком
    return clean_text
```

```python
processed_text = remove_tags(text)
print(processed_text)
```

## 4. Pymorphy vs Mystem

```python
from pymorphy2 import MorphAnalyzer
from pymystem3 import Mystem

mystem = Mystem()
mystem_analysis = [record for record in mystem.analyze(processed_text) if record['text'].isalnum()] #убираем пробелы и пунктуацию
tokens = [record['text'] for record in mystem_analysis]
mystem_lemmas = []
for record in mystem_analysis:
    try:
        mystem_lemmas.append(record['analysis'][0]['lex'])
    except (KeyError, IndexError):
        mystem_lemmas.append(record['text'])
        
pymorphy = MorphAnalyzer()
pymorphy_analysis = [pymorphy.parse(token) for token in tokens]
pymorphy_lemmas = [token[0].normal_form for token in pymorphy_analysis]

print(f'{sum(1 for i in range(len(tokens)) if mystem_lemmas[i] != pymorphy_lemmas[i])} '
      f'differences in {len(tokens)} tokens\n')
for i in range(len(tokens)):
    mystem_lemma = mystem_lemmas[i].lower().replace('ё', 'е')
    pymorphy_lemma = pymorphy_lemmas[i].lower().replace('ё', 'е')
    # при сравнении лемм будем игнорировать регистр и разницу е/ё
    if mystem_lemma != pymorphy_lemma:
        print(f'MyStem Lemma - {mystem_lemma}\nPyMorphy Lemma - {pymorphy_lemma}\n'
              f'Context - {" ".join(tokens[max(0, i - 4):min(len(tokens), i + 5)])}\n')
```

В каких-то случаях MyStem работает лучше (возможно, из-за того, что он принимает на вход целое предложение и анализирует контекст, чего нет в PyMorphy). Это видно, например, в этом примере, где PyMorphy допустил явную ошибку:

MyStem Lemma - данные<br>
PyMorphy Lemma - дать<br>
Context - пользователей и выводить эти данные в различных формах инфракрасная<br>

PyMorphy не знал, в каком контексте употреблялось слово "данные" и счёл наиболее вероятной леммой самую часто встречающуюся.

С другой стороны, в каких-то случаях, наоборот, ошибается MyStem, например,

MyStem Lemma - он<br>
PyMorphy Lemma - они<br>
Context - где расположить В этом им поможет Сумасшедшее яйцо Также<br>

Но всё же большая часть различий объясняется той или иной трактовкой языковых единиц, например:

MyStem Lemma - много<br>
PyMorphy Lemma - большой<br>
Context - сайте которые полезны пользователям больше всего Больше не нужно<br>

Здесь MyStem как будто "увидел", что "больше" входит в состав словосочетания "больше всего", а у PyMorphy такой возможности не было, он расценивал "больше" как отдельное слово. Какой из подходов правильный - зависит от задачи.

```python
tokens = []
def analyze_with_mystem_instance(mystem_instance: Mystem) -> list:
    analysis = [record for record in mystem_instance.analyze(processed_text) if
                record['text'].isalnum()]  # убираем пробелы и пунктуацию
    global tokens
    if not tokens:
        tokens = [record['text'] for record in analysis]
    lemmas = []
    for record in analysis:
        try:
            lemmas.append(record['analysis'][0]['lex'])
        except (KeyError, IndexError):
            lemmas.append(record['text'])
    return lemmas
```

```python
non_disambig_lemmas = analyze_with_mystem_instance(Mystem(disambiguation=False))
disambig_lemmas = analyze_with_mystem_instance(Mystem(disambiguation=True))
print(f'{sum(1 for i in range(len(non_disambig_lemmas)) if non_disambig_lemmas[i] != disambig_lemmas[i])} '
      f'differences in {len(tokens)} tokens\n')
for i in range(len(tokens)):
    if non_disambig_lemmas[i] != disambig_lemmas[i]:
        print(f'Non-disambig Lemma - {non_disambig_lemmas[i]}\nDisambig Lemma - {disambig_lemmas[i]}\n'
              f'Context - {" ".join(tokens[max(0, i - 4):min(len(tokens), i + 5)])}\n')
```
